#### 常用5種視覺統計圖形

1. 直方圖 Histogram: matplotlib.pyplot.hist()
2. 長條圖 Bar Plot : matplotlib.pyplot.bar()
3. 線圖 Line Plot  : matplotlib.pyplot.plot()
4. 散佈圖 Scatter Plot: matplotlib.pyplot.scatter()
5. 盒鬚圖 Box Plot : matplotlib.pyplot.boxplot()
6. 圓餅圖 Pie Plot : matplotlib.pylot.pie()

直方圖: 
1. 用於表示連續性資料，表達一組資料集(dataset)的大致樣態: time-series: daily, annual...
2. 主要表達資料分布結果: 連續時間區間內，最大分布
3. 橫軸變數(X軸): 數值行連續變數(time-series 0-23)

長條圖: 
1. 用於統計不同類型資料的統計值: groups: 分類與其資料: age group...
2. 橫軸變數(X軸): 類別形離散變數

線圖:
1. 資料具順序性

盒鬚圖:
1. 顯示最大值和最小值以及中位數
2. 顯示資料是否集中等訊息

```python
#1.直方圖:
import matplotlib.pyplot as plt 
import random

hourlypoweruse = [random.randint(1, 15) for _  in range(24)]
print(hourlypoweruse)

plt.hist(hourlypoweruse, range(24))
plt.xlabel("Hours")
plt.ylabel("Power Consumation")
plt.title("Hourly Power Consumation per day Histogram")
plt.show()
```

```python
import collections
import matplotlib.pyplot as plt

age = [22, 55, 62, 45, 21, 22, 34, 42, 42, 4, 99, 102, 110, 120, 121, 122, 130, 111, 115, 112, 80, 75, 65, 54, 44, 43, 42, 48]
t = collections.Counter(age)
t = sorted(t.items())
print(t)

bins = [i*15 for i in range(1,28)] #區間 數量控制
print(bins)

#bottom: 縱軸從_開始(最下方)
#density = True: 比例方式表示
#density = False: 以資料數字直接顯示
#hihsttype : default:bar
#histtype = 'Bar'
#histtype = 'Barstacked'
#histtype = 'step'
#histype = 'stepfill'
#align對其 = 'left', 'mid', 'right'
#log: log scale
#orientation: 'vertical', 'horizontal'
#color:顏色設定

plt.hist(age, bins, bottom = 0, density = False, histtype = 'bar', align = 'right', log = False, orientation = "horizontal", color='magenta')
plt.show()
```

```python
#2. 線圖
import matplotlib.pyplot as plt #代稱plt

data = {'Eva':4, 'Max':2, 'John':6, 'Chris':11, 'Bryan':8, 'Jenny':1}
mylist = data.items()
mylist = sorted(mylist, key = lambda x:x[1], reverse=True)
x, y = zip(*mylist) #平行assignment


plt.plot(x, y)
plt.xlabel("Name")
plt.ylabel("Amount")
plt.title("Name")
```

```python
#3.長條圖

import matplotlib.pyplot as plt #... as plt: 代稱為plt

data = {'Eva':4, 'Max':2, 'John':6, 'Chris':11, 'Bryan':8, 'Jenny':1}
data = sorted(data.items(), key = lambda x:x[1], reverse = True)
x, y = zip(*data)
print(x, y)

plt.bar(x, y)
plt.show()
```

```python
#4.散佈圖
#顯示大略請向
import matplotlib.pyplot as plt
import random

speed = [random.randint(1, 25) for _ in range(30)]
dist = [random.randint(1, 120) for _ in range(30)]
print(speed)
print(dist)

plt.scatter(speed, dist)
plt.show()
plt.plot(speed, dist)
plt.show()
```

```python
#5.盒鬚圖
import matplotlib.pyplot as plt
import random

data = [random.randint(20, 30) for _ in range(30)]
data2 = [random.randint(10, 100) for _ in range(30)]
data3 = [random.randint(50, 100) for _ in range(30)]
print(data)

plt.boxplot([data, data2, data3])
plt.show()

#Orange Line: median
```

```python
#6.Pie Chart
import matplotlib.pyplot as plt

labels = ['a', 'b', 'c', 'd', 'e', 'f']
data = [33, 52, 12, 17, 62, 48]

#計算比例
pct = list(map(lambda x:(x/sum(data))*100, data))
pct = list(map(lambda x: round(x, 2), pct))

plt.pie(data, labels = labels, autopct = '%5.2f%%')
plt.title("Pie Chart for Alphabet")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = [
    [3, 25, 50, 20],
    [40, 23, 51, 72],
    [35, 22, 45, 19]
]

x = np.arange(len(data[0])) #X刻度
fig = plt.figure(figsize=[5,4], facecolor='paleturquoise', edgecolor='teal') #製作一個空白畫布
ax = fig.add_axes([0, 0, 1, 1]) #from 0 to 1 
ax.bar(x+0.0, data[0], color='lightseagreen', width=0.25)
ax.bar(x+0.25, data[1], color='mediumturquoise', width=0.25) #資料刻度x+0.25
ax.bar(x+0.5, data[2], color='darkslategray', width=0.25)

plt.title('Data with Categories A-D')
ax.set_ylabel('Count')
ax.set_xlabel('Data')
ax.legend(labels=['data[0]', 'data[1]', 'data[2]'])
plt.xticks(x, ['A', 'B', 'C', 'D']) #將0, 1, 2, 3 改成A,B,C,D
plt.show()
```

## ax V.S. plt

matplotlib has a module called pyplot
pyplot has a module called figure
figure has ax among other things

```python
import numpy as np
import matplotlib.pyplot as plt

men = [10, 20, 25, 55, 27]
women = [35, 32, 34, 20, 25]
x = np.arange(len(men))

fig = plt.figure(figsize=[5,4], facecolor='peachpuff', edgecolor='black') #製作一個空白畫布
ax = fig.add_axes([0, 0, 1, 1])
ax.bar(x, men, color='indianred', width=0.4)
ax.bar(x, women, color='darkorange', bottom=men, width=0.4) #bottom: 從men再加0.25
ax.legend(labels=['Men', 'Women']) #注意順序
ax.set_ylabel('People')
ax.set_xlabel('Height')
plt.yticks(np.arange(0, 10, 101))
plt.xticks(x, ['A', 'B', 'C', 'D', 'E'])
plt.show()
#找出總和之間的差異與極大差異值
```

```python
import numpy as np
import matplotlib.pyplot as plt

men = [20, 25, 30, 35, 27]
sd_m = [2, 1, 3, 2, 4]  #標準差
women = [25, 32, 34, 20, 25]
errorw = [3, 1, 2, 2, 1]

#(fig, ax) = plt.subplots() #create a figure and a set of subplots
x = np.arange(5)
bw = 0.35
opacity = 0.6 #不透明度

error_config = {'ecolor':"0.1"} #誤差值
rec1 = plt.bar(x, men, width=bw, alpha=opacity, color='b', yerr=sd_m, error_kw=error_config, label='Men')
rec2 = plt.bar(x+bw, women, width=bw, alpha=0.4, color='r',yerr=errorw, error_kw=error_config, label='Women')

plt.xlabel('Group')
plt.ylabel('Score')
plt.title('Group Score')
plt.xticks(x+0.16, ['A', 'B', 'C','D', 'E'])
#plt.tight_layout() 圖形便扁平
plt.savefig('1.png') #圖形輸出。處存之前不可用plt.show
plt.show()
```
